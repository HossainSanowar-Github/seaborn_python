# Create Relation plot between two data
* using relplot
* using "hue" for multiple color code
* using "style" for multiple shape
* using "size" and "sizes" for compare data value

1. How to import data using pandas
2. How to create first scatter chart with seaborn
3. Improving charts with color coding, style and size

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
ls
```

```python
data=pd.read_excel('sales.xlsx')
df=data.copy()
```

```python
df.head()
```

# Create the relation plot
* Relationship plot between Sales and Profit

```python
sns.relplot('Sales','Profit',data=df)
```

# using "hue" for multiple color

```python
sns.relplot('Sales','Profit',hue='Order Priority',data=df)
```

# Using "style" for different shape mode

```python
sns.relplot('Sales','Profit',hue='Order Priority',style='Order Priority',data=df)
```

```python
sns.relplot('Sales','Profit',hue='Order Priority',style='Ship Mode',data=df)
```

# using "size" for compare

```python
sns.relplot('Sales','Profit',size='Discount',data=df)
```

```python
sns.relplot('Sales','Profit',data=df,size='Discount',sizes=(50,200))
```
