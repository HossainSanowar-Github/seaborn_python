# Scater plot on categorical data
* Create "catplot"
* using "query" method
* create "swarm" plot

1. How to plot scatter plot or swarm plot on categorical data
2. How to remove magnitude of swarm plot
3. How to query data within chart parameters

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_excel('sales.xlsx')
df=data.copy()
df.head(3)
```

# Create Scater plot on categorical data

```python
sns.catplot('Ship Mode','Sales',data=df)
```

# Using Jitter

```python
sns.catplot('Ship Mode','Sales',data=df,jitter=False)
```

# Using "query" method 

```python
sns.catplot('Ship Mode','Sales',data=df.query('Profit>2000'),jitter=False)
```

# Using "swarm" chart

```python
sns.catplot('Ship Mode','Sales',data=df.query('Profit>2000'),kind="swarm")
```

# Using as vertical

```python
sns.catplot('Sales','Ship Mode',data=df.query('Profit>2000'),kind="swarm")
```
