# Create distplot
* using kde, hist, rug, bins
* using condition
* Create KDE plot

1. How to distribution data using distplot
2. How to take an impact of another categorical dimension
3. How to create multiple chart by looping over dimension values

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_excel('sales.xlsx')
df=data.copy()
df.head(3)
```

```python
sns.distplot(df.Sales)
```

```python
sns.distplot(df.Profit)
```

```python
sns.distplot(df.Profit,kde=False)
```

```python
sns.distplot(df.Profit,hist=False)
```

```python
sns.distplot(df.Profit,hist=False,rug=True)
```

```python
sns.distplot(df.Profit,hist=False,rug=True,bins=100)
```

# using condition

```python
sns.distplot(df.Profit[(df.Region=="West")&(df.Profit>0)],hist=False,rug=True,bins=100)
```

# Create kde plot

```python
sns.kdeplot(df.Profit,bw=100)
```
