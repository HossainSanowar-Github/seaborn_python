# Create Line chart
* change the date format to 'year'
* using "kind" method
* using estimator, ci, hue, style, col, col_wrap

1. How plot data on dates
2. How to create first line chart with seaborn
3. Improving charts with color coding, style and size
4. Extracting year from Dates
5. Plotting confidence intervals or standard deviation around the mean

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_excel("sales.xlsx")
df=data.copy()
```

```python
df.head()
```

# Create line plot

```python
sns.relplot("Order Date","Sales",data=df,kind='line')
```

# Extract year from Order Date

```python
df['year']=df['Order Date'].dt.year
```

```python
df.head(3)
```

```python
df.year.unique()
```

```python
sns.relplot('year','Sales',data=df,kind='line')
```

```python
#remove confidence interval
sns.relplot('year','Sales',data=df,kind='line',ci=None)
```

```python
#confidence interval as "sd"
sns.relplot('year','Sales',data=df,kind='line',ci='sd')
```

```python
#using "estimator"
sns.relplot('year','Sales',data=df,kind='line',estimator=None)
```

```python
sns.relplot('year','Sales',data=df,kind='line',estimator=sum)
```

```python
#using "hue" for different color
sns.relplot('year','Sales',data=df,kind='line',hue='Ship Mode')
```

```python
#using "hue" for different color and "style" for different shape
sns.relplot('year','Sales',data=df,kind='line',hue='Ship Mode',style='Product Category')
```

```python
#using "hue" for different color and "style" for different shape,and confidence interval
sns.relplot('year','Sales',data=df,kind='line',hue='Ship Mode',style='Product Category',ci=None)
```

```python
#using "hue" for different color, ci, and col
sns.relplot('year','Sales',data=df,kind='line',hue='Ship Mode',col='Product Category',ci=None)
```

```python
#using "hue" for different color,col,ci
sns.relplot('year','Sales',data=df,kind='line',hue='Ship Mode',col='Region',ci=None)
```

```python
#using "hue" for different color, col, ci,col_wrap
sns.relplot('year','Sales',data=df,kind='line',hue='Ship Mode',col='Region',ci=None,col_wrap=4)
```
