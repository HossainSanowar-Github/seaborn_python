```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_excel("sales.xlsx")
df=data.copy()
```

```python
df.head(3)
```

# Using coloum

```python
sns.relplot('Sales','Profit',data=df,col='Customer Segment',col_wrap=2)
```

```python
sns.relplot('Sales','Profit',data=df,col='Customer Segment',col_wrap=2,hue='Order Priority')
```

```python
sns.relplot('Sales','Profit',data=df,col='Customer Segment',col_wrap=2,hue='Ship Mode')
```

# Using Multiple row

```python
sns.relplot('Sales','Profit',data=df,col='Customer Segment',row='Product Category',hue='Ship Mode')
```
