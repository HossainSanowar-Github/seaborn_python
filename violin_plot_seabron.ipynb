# Violin Chart
* using catplot

1. How to plot violin chart using catplot
2. How to take an impact of another categorical dimension
3. How to create multiple chart by looping over dimension values

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_excel('sales.xlsx')
df=data.copy()
df.head(3)
```

# Violin chart

```python
sns.catplot(x='Product Category',y='Shipping Cost',kind="violin",data=df)
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',kind="violin",hue="Ship Mode",data=df)
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',kind="violin",hue="Ship Mode",inner="stick",data=df)
```
