1. How to pointplot data using catplot
2. How to take an impact of another categorical dimension
3. How to create multiple chart by looping over dimension values

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_excel('sales.xlsx')
df=data.copy()
df.head(3)
```

# Create point plot

```python
sns.catplot(x="Product Category",y='Sales',data=df,kind="point")
```

```python
sns.catplot(x="Product Category",y='Sales',hue="Ship Mode",data=df,kind="point")
```

```python
sns.catplot(x="Product Category",y='Sales',hue="Ship Mode",col="Customer Segment",data=df,kind="point")
```
